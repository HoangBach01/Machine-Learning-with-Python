```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
import tensorflow as tf
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(60000, 784)
print(X_train.shape)
X_test = X_test.reshape(10000, 784)
print(X_test.shape)
```

```python
def svm_mnist(X_train, y_train, X_test, y_test, use_pca=False):
    from sklearn.svm import SVC
    import time
    tic = time.time()
    if use_pca:
        from sklearn.decomposition import PCA
        pca = PCA(use_pca)
        X_train = pca.fit_transform(X_train)
        X_test = pca.transform(X_test)
    model = SVC()
    model.fit(X_train, y_train)
    print('Accuracy train: {:.4f}'.format(model.score(X_train, y_train)))
    toc_1 = time.time()
    print('Training time: {:.2f} seconds'.format(toc_1-tic))
    print('Accuracy test : {:.4f}'.format(model.score(X_test, y_test)))
    toc_2 = time.time()
    print('Predict time: {:.2f} seconds'.format(toc_2-toc_1))
```

```python
svm_mnist(X_train, y_train, X_test, y_test, use_pca=0.95)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, BatchNormalization, Input, ReLU, Flatten
from tensorflow.keras.models import Model
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

input = Input(shape=(28, 28))
x = Flatten()(input)
x = Dense(300)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
x = Dense(200)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
x = Dense(200)(x)
x = BatchNormalization()(x)
x = ReLU()(x)
output = Dense(10, activation='softmax')(x)
model = Model(inputs=input, outputs=output)

model.compile(loss=keras.losses.SparseCategoricalCrossentropy(), 
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50)
model.evaluate(X_test, y_test)
```
