```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import pandas_profiling as pp
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

%matplotlib inline
```

```
text = [
        'ngay hom nay la ngay may',
        'toi mot gio rang chiu',
        'ong gia di nhanh qua',
]

count_vec = CountVectorizer()
count_vec.fit(text)
bag_of_words = count_vec.transform(text)
print(bag_of_words.toarray())
```

```
X = np.array([["Red", "Sports", "Domestic"],
              ["Red", "Sports", "Domestic"], 
              ["Red", "Sports", "Domestic"], 
              ["Yellow", "Sports", "Domestic"], 
              ["Yellow", "Sports", "Imported"],
              ["Yellow", "SUV", "Imported"], 
              ["Yellow", "SUV", "Imported"], 
              ["Yellow", "SUV", "Domestic"],
              ["Red", "SUV", "Imported"], 
              ["Red", "Sports", "Imported"]])
y = np.array(["Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No", "No", "Yes"])
X = pd.DataFrame(X, columns=['Color', 'Type', 'Origin'])
X = pd.get_dummies(X)
y = np.array(pd.get_dummies(y, drop_first=True)).flatten()
y
```

```
X
```

```
model = BernoulliNB()
model.fit(X, y)
model.score(X, y)
```

```
X_test = [[0, 1, 0, 1, 0, 1]]
prediction = model.predict(X_test)[0]
print('Probability of Stolen: {:.4f}'.format(model.predict_proba(X_test)[0,1]))
print('Prediction: {}'.format('Stolen' if prediction==1 else 'Not stolen'))
```

```
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X, y)
model.score(X, y)
X_test = [[0, 1, 0, 1, 0, 1]]
prediction = model.predict(X_test)[0]
print('Probability of Stolen: {:.4f}'.format(model.predict_proba(X_test)[0,1]))
print('Prediction: {}'.format('Stolen' if prediction==1 else 'Not stolen'))
```

```
import pickle
pkl_filename = 'model.pkl'
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)
```
